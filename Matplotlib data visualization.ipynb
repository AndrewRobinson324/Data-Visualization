```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional
plt.plot(x,y,)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented 

#Creates a figure
fig = plt.figure()

# Add a set of axes to it
axes = fig.add_axes([0.1,0.1,0.8,0.8])
# Then plot of these axes
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title("Set Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
axes1.set_title('Larger Plot')
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
#axes.plot(x,y)
#for current_ax in axes:
   # current_ax.plot(x,y)
    
# Takes care of any overlapping
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1, figsize = (8,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi = 200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,0.9])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'X Cubed')

#Add in a legend
# need a label
ax.legend(loc = 2)
```

```python
#plot appearance

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, color = 'purple', linewidth = 5, alpha = 0.5) # or use RGB hex code



```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color = 'purple',lw = 1,ls = '-',marker = 'o',markersize = 15,markerfacecolor = 'yellow',markeredgecolor = 'green')

```
